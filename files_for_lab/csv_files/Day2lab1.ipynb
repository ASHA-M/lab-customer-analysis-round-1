```python
import pandas as pd
import numpy as np

file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')

column_names1 = file1.columns
column_names2 = file2.columns
column_names3 = file3.columns

#display(column_names1)
#display(column_names2)
#display(column_names3)

file3 = file3.rename(columns={'State':'ST'})
display(column_names3)


data = pd.concat([file1, file2, file3], axis=0) 
data.shape
data
```

```python
data._get_numeric_data()
```

```python
columns = []
for i in data.columns:
    columns.append(i.lower())
data.columns = columns
```

```python
data.head()
```

```python
data = data.drop(['education', 'number of open complaints'], axis=1)
```

```python
data.info()
```

```python
data['customer lifetime value'] = data['customer lifetime value'].astype('str')
data['customer lifetime value'] = data['customer lifetime value'].apply(lambda x: x.replace('%', ''))
data['customer lifetime value'] = data['customer lifetime value'].astype('float')
```

```python
data['customer lifetime value'] = data['customer lifetime value'] / 10000
data.head()
```

```python
data = data.drop_duplicates()
```
